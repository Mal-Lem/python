<b> Colonnes </b>

- Country: Nom du pays
- Density (P/Km2): Densité de population mesurée en personnes par kilomètre carré
- Agricultural Land (%): Pourcentage de la surface des terres utilisées à des fins agricoles
- Land Area (Km2): Superficie totale des terres du pays en kilomètres carrés
- Armed Forces Size: Taille des forces armées dans le pays
- Birth Rate: Nombre de naissances pour 1 000 habitants par an
- CO2 Emissions: Émissions de dioxyde de carbone en tonnes

```python
import numpy as np

url = '/content/world-2023.csv'

data = np.genfromtxt(url,delimiter=',',encoding='utf-8',dtype=str)
data

#import pandas as pd
#data = pd.read_csv(url)
#data
```

```python
entete = data[0]
entete
```

```python
pays = data[1:,0]
pays
```

```python
donnees_numeriques = data[1:,1:]
donnees_numeriques = np.char.replace(donnees_numeriques,'%','')
donnees_numeriques = donnees_numeriques.astype(float)
donnees_numeriques
```

```python
nan_mask = np.isnan(donnees_numeriques)

countries_with_no_data = []

for i, row in enumerate(nan_mask):
    if np.all(row):
        countries_with_no_data.append(i)

print("pays avec aucune donnée numérique:", countries_with_no_data)

nan_percentage = np.sum(nan_mask, axis=0) / donnees_numeriques.shape[0] * 100

for i, percentage in enumerate(nan_percentage):
    print(f"colonne {i}: {percentage:.2f}% de valeurs NaN")
```

```python
donnees_numeriques = np.nan_to_num(donnees_numeriques, nan=0)
donnees_numeriques
```

```python
nan_percentage = np.sum(nan_mask, axis=0) / donnees_numeriques.shape[0] * 100

for i, percentage in enumerate(nan_percentage):
    print(f"colonne {i}: {percentage:.2f}% de valeurs NaN")
```
